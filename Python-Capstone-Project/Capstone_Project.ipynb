```python
# TASK 2: Replace missing values with running average using a for loop

# Step 2a: Convert the 'Cost' column to numeric, forcing the 'NaN' strings to actual numeric NaN
ProjectDF['Cost'] = pd.to_numeric(ProjectDF['Cost'], errors='coerce') 

# Initialize variables for calculating the running average
running_sum = 0.0 # Initialize sum of valid costs
running_count = 0 # Initialize count of valid costs

# Step 2b: Loop through each row of the DataFrame
for index, row in ProjectDF.iterrows():
    cost = row['Cost']
    
    # Check if the cost value is missing (NaN)
    if pd.isna(cost):
        # If a valid cost has been seen before (count > 0)
        if running_count > 0:
            # Calculate the current running average
            current_running_avg = running_sum / running_count
            
            # Use .loc to replace the missing value in the DataFrame
            ProjectDF.loc[index, 'Cost'] = current_running_avg
            print(f"Replaced NaN at Project: {row['Project']} with running average: {current_running_avg:.2f}")
            
        else:
            # If the very first value is NaN, we cannot calculate an average yet.
            print(f"Skipping replacement for {row['Project']}: No valid costs encountered yet.")
            
    else:
        # If the cost is a valid number, update the running sum and count
        running_sum += cost
        running_count += 1
        
print("\n--- ProjectDF After Task 2 ---")
print(ProjectDF)
```

```python
import pandas as pd
import numpy as np

# --- 1. THE FIX: Set Pandas option for clean numerical display ---
# This applies the fix you asked for: showing large float numbers with 2 decimal places 
# instead of scientific notation (e.g., 1002000.00 instead of 1.002e+06)
pd.set_option('display.float_format', '{:.2f}'.format) 

# --- TASK 6: Demote and Delete Records ---

# Load the FinalDF updated after Task 5 (This is where the Cost column is located)
# We use the previous save point: Final_Updated_Task5.csv
FinalDF = pd.read_csv('Final_Updated_Task5.csv')

# 6a. Identify IDs of employees with "Failed" projects
failed_project_ids = FinalDF[FinalDF['Status'] == 'Failed']['ID'].unique()
print(f"IDs with Failed Projects (to be demoted): {failed_project_ids}")

# Create a mapping of current Designation Levels (one entry per employee ID)
Designation_Map = FinalDF[['ID', 'Designation Level']].drop_duplicates().set_index('ID')['Designation Level'].to_dict()

# Apply Demotion: Level increases by 1 for employees with failed projects
for emp_id in failed_project_ids:
    if emp_id in Designation_Map:
        current_level = Designation_Map[emp_id]
        # Demote: Level increases by 1 (e.g., 2 -> 3)
        Designation_Map[emp_id] = current_level + 1
        print(f"  Demoting {emp_id}: {current_level} -> {current_level + 1}")

# Update the 'Designation Level' column in the main FinalDF using the new map
FinalDF['Designation Level'] = FinalDF['ID'].map(Designation_Map)


# 6b. Delete employees whose Designation Level is now above 4
# Rule: "If anyone crosses 4 then he loses eligibility"
ineligible_ids = FinalDF[FinalDF['Designation Level'] > 4]['ID'].unique()

# Filter FinalDF to exclude rows belonging to ineligible IDs
FinalDF = FinalDF[~FinalDF['ID'].isin(ineligible_ids)]

print(f"\nIDs Deleted (Designation Level > 4): {ineligible_ids}")
print(f"FinalDF records remaining: {len(FinalDF)}")

print("\n--- FinalDF After Task 6 (Demotion and Deletion) ---")
# The Cost values will now be displayed cleanly!
print(FinalDF[['ID', 'First Name', 'Designation Level', 'Cost', 'Status']].to_markdown(index=False))

# Save the Final DataFrame
FinalDF.to_csv('Final_Updated_Task6.csv', index=False)
```
